```python
# Goal: create and graph looping iterations of the Collatz conjecture (3n+1)

from random import randint
from matplotlib import pyplot as plt
```

```python
# Plot a single random iteration

d = randint(1,1000000) # starting integer
nlist = []

while len(nlist)<1000 and d != 1:
    nlist.append(d)
    if d % 2 == 0:
        d = d // 2
    else:
        d = (d * 3) + 1

plt.plot(nlist)
plt.show()
print(nlist)
```

```python
# Plot multiple iterations together

xmin = 100
xmax = 150

d = range(xmin,xmax) # starting integer
x = 0
nlist = [[] for i in range(xmax-xmin)]

for i in d:
    while i != 1:
        nlist[x].append(i)
        if i == 2:
            x += 1
        if i % 2 == 0:
            i = i // 2
        else:
            i = i * 3 + 1

            
#print(nlist)

z = range(0,xmax-xmin)

for y in z:
    plt.plot(nlist[y])
    #plt.show()

```

```python
# Plot multiple iterations separately

xmin = 100
xmax = 110

d = range(xmin,xmax) # starting integer
x = 0
nlist = [[] for i in range(xmax-xmin)]

for i in d:
    while i != 1:
        nlist[x].append(i)
        if i == 2:
            x += 1
        if i % 2 == 0:
            i = i // 2
        else:
            i = i * 3 + 1

            
print(nlist)

z = range(0,xmax-xmin)

for y in z:
    plt.plot(nlist[y])
    plt.show()

```
